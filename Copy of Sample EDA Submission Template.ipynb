## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#filled all columns containing NaN with 0
columns_to_fill_zero = ['Championships', 'Race_Wins', 'Pole_Positions', 'Podiums', 'Fastest_Laps','Championship Years']
df[columns_to_fill_zero] = df[columns_to_fill_zero].fillna(0)

#created a new column called pole_to_win_rate to see how many poles did the driver convert to wins
df['Pole_To_Win_Rate'] = np.where(df['Pole_Positions'] > 0, df['Race_Wins'] / df['Pole_Positions'], 0)
#created a new column podiums per season to see how many podiums a driver scored in a season for better insight
df['Podiums_Per_Season'] = np.where(df['Years_Active'] > 0, df['Podiums'] / df['Years_Active'], 0)
df.head()
````

### What all manipulations have you done and insights you found?

### Data Manipulations Performed:
1. **Imputation of Missing Values:** Replaced `NaN` values in core performance statistical columns (`Championships`, `Race_Wins`, etc.) with `0`, as missing entries historically imply the driver never achieved that specific milestone.
2. **Feature Engineering (Efficiency Indices):** * Created `Pole_To_Win_Rate` to measure a driver's Sunday execution dominance when starting from the front.
    * Created `Podiums_Per_Season` to normalize carrier yields against total active career duration.

### Initial Wrangling Insights:
* Our newly engineered features (`Pole_To_Win_Rate` and `Podiums_Per_Season`) provide the foundation required to evaluate driver "bang-for-your-buck" value, preparing us directly to address the core business objective.

## ***4. Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables***

#### Chart - 1

````python
# Chart - 1 visualization code
top_10_winners = df.sort_values(by = 'Race_Wins' , ascending = False).head(10)

plt.figure(figsize=(12, 6))

colors = sns.color_palette("coolwarm", n_colors=10)

ax = sns.barplot(
    x='Race_Wins',
    y='Driver',
    data=top_10_winners,
    palette=colors,
    hue='Driver',
    legend=True
)

plt.title('Top 10 Most Successful Formula 1 Drivers by Absolute Race Wins', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Total Career Race Wins', fontsize=12)
plt.ylabel('Driver Name', fontsize=12)
plt.show()
````

##### 1. Why did you pick the specific chart?

I chose a horizontal bar chart instead of a vertical one because the names of the drivers can be quite long. This layout keeps the Y-axis clear and easy to read. Also, since there are hundreds of historical drivers, I focused on the top 10 to showcase the best performers without overcrowding the visualization.

##### 2. What is/are the insight(s) found from the chart?

As shown on the graph, there is an pattern in terms of the historical performance of the Formula 1 drivers. The leaders of the field have a very large number of wins, with both Lewis Hamilton and Michael Schumacher having won over 90 races. The gap from one position to another among the top 10 drivers is very sharp. Showing how rare it is for a driver to achieve long term race dominance across their career

````python
#Setup the figure size and theme
plt.figure(figsize=(12, 6))
sns.set_style("whitegrid")

#Plotted a crisp histogram using exact counts
sns.histplot(
    data=df,
    x='Years_Active',
    bins=19,
    color='firebrick',
    edgecolor='black',
    alpha=0.85
)

#Clean labels and analytical title
plt.title('Distribution of F1 Driver Career Lifespans (Total Years Active)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Number of Years Active in Formula 1', fontsize=12)
plt.ylabel('Number of Drivers', fontsize=12)

#Set absolute tick marks for every year value from 1 to 19
plt.xticks(range(1, 20))

plt.tight_layout()
plt.show()
````

##### 1. Why did you pick the specific chart?

The regular histogram is the right choice since it categorizes each data into distinct year intervals. We can thus easily observe the exact number of drivers that have driven for each number of years

##### 2. What is/are the insight(s) found from the chart?

There is a strong right skewness in the data. The overwhelming majority of F1 drivers in history have their careers limited to between 1 and 3 years in the sport. It is rare to find a driver who has lasted more than 5 years in F1, while a select few hyper-elite veteran drivers have been able to have 10 to 19 years long careers.
