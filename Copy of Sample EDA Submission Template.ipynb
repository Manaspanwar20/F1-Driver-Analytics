## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#filled all columns containing NaN with 0
columns_to_fill_zero = ['Championships', 'Race_Wins', 'Pole_Positions', 'Podiums', 'Fastest_Laps','Championship Years']
df[columns_to_fill_zero] = df[columns_to_fill_zero].fillna(0)

#created a new column called pole_to_win_rate to see how many poles did the driver convert to wins
df['Pole_To_Win_Rate'] = np.where(df['Pole_Positions'] > 0, df['Race_Wins'] / df['Pole_Positions'], 0)
#created a new column podiums per season to see how many podiums a driver scored in a season for better insight
df['Podiums_Per_Season'] = np.where(df['Years_Active'] > 0, df['Podiums'] / df['Years_Active'], 0)
df.head()
````

### What all manipulations have you done and insights you found?

### Data Manipulations Performed:
1. **Imputation of Missing Values:** Replaced `NaN` values in core performance statistical columns (`Championships`, `Race_Wins`, etc.) with `0`, as missing entries historically imply the driver never achieved that specific milestone.
2. **Feature Engineering (Efficiency Indices):** * Created `Pole_To_Win_Rate` to measure a driver's Sunday execution dominance when starting from the front.
    * Created `Podiums_Per_Season` to normalize carrier yields against total active career duration.

### Initial Wrangling Insights:
* Our newly engineered features (`Pole_To_Win_Rate` and `Podiums_Per_Season`) provide the foundation required to evaluate driver "bang-for-your-buck" value, preparing us directly to address the core business objective.

````python
active_drivers = df[df['Active'] == True].sort_values(by='Race_Entries', ascending=False).copy()

#Set up the plotting foundations
fig, ax1 = plt.subplots(figsize=(14, 7))
x_indices = np.arange(len(active_drivers))
bar_width = 0.35

#Plot the first set of bars: Race Entries (Left Y-Axis)
bars1 = ax1.bar(
    x_indices - bar_width/2,
    active_drivers['Race_Entries'],
    width=bar_width,
    color='#1f77b4',
    label='Race Entries (Experience)'
)
ax1.set_xlabel('Active Drivers', fontsize=12, fontweight='bold')
ax1.set_ylabel('Total Career Race Entries', color='#1f77b4', fontsize=12, fontweight='bold')
ax1.tick_params(axis='y', labelcolor='#1f77b4')
ax1.set_xticks(x_indices)
ax1.set_xticklabels(active_drivers['Driver'], rotation=45, ha='right')

#Create the secondary axis sharing the same x-axis
ax2 = ax1.twinx()

#Plot the second set of bars: Points Per Entry (Right Y-Axis)
bars2 = ax2.bar(
    x_indices + bar_width/2,
    active_drivers['Points_Per_Entry'],
    width=bar_width,
    color='#ff7f0e',
    label='Points Per Entry (Efficiency)'
)
ax2.set_ylabel('Average Points Per Race Entry', color='#ff7f0e', fontsize=12, fontweight='bold')
ax2.tick_params(axis='y', labelcolor='#ff7f0e')

#Add exact values directly on top of the bars for crisp detail without clutter
ax1.bar_label(bars1, padding=3, fontsize=9, fmt='%g')
ax2.bar_label(bars2, padding=3, fontsize=9, fmt='%.1f')

#Add titles and clean up layout
plt.title('Active Driver Profiles: Total Career Experience vs. Point Scoring Efficiency', fontsize=14, fontweight='bold', pad=20)
fig.tight_layout()
plt.show()
````

````python
# Chart - 11 visualization code
active_fastest = df[df['Active'] == True].copy()

fastest_lap_col = next((col for col in df.columns if 'fastest' in col.lower()), None)
starts_col = next((col for col in df.columns if 'start' in col.lower() or 'entry' in col.lower()), None)

if not fastest_lap_col: fastest_lap_col = 'Fastest_Laps'  
if not starts_col: starts_col = 'Race_Starts'

active_fastest = active_fastest[active_fastest[starts_col] > 0]

active_fastest['Fastest_Lap_Rate'] = (active_fastest[fastest_lap_col] / active_fastest[starts_col]) * 100

active_fastest = active_fastest.sort_values(by='Fastest_Lap_Rate', ascending=False)

num_drivers = len(active_fastest)
plt.figure(figsize=(12, max(6, num_drivers * 0.4)))
sns.set_theme(style="whitegrid")

ax = sns.barplot(
    x='Fastest_Lap_Rate',
    y='Driver',
    data=active_fastest,
    palette='coolwarm', 
    hue='Driver',
    legend=False
)

for container in ax.containers:
    ax.bar_label(container, fmt='%.1f%%', padding=5, fontweight='bold', color='#333333')

plt.title('Raw Pace Profiles: Career Fastest Lap Conversion Rates (Active Grid)', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Percentage of Races with a Fastest Lap (%)', fontsize=12)
plt.ylabel('Active Driver Name', fontsize=12)

plt.tight_layout()
plt.show()
````
